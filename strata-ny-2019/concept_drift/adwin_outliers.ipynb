```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from skmultiflow.drift_detection.adwin import ADWIN
from numpy import genfromtxt
data_stream = genfromtxt('Power.csv', delimiter=',')
import pandas as pd 

data_pd = pd.read_csv("Power.csv") 
data_pd.columns = ["date","power"]
print(data_pd.head())
print(data_pd.dtypes)
data_pd['date']=pd.to_datetime(data_pd['date'],unit='ms')
print(data_pd.head(20))
tolerance=100
sns.set(rc={'figure.figsize':(25, 4)})
        
plt.plot(data_pd['date'], data_pd['power'])
plt.show()

data = data_stream[:,1]
outliers = []
clean_data = []

input_data_std = 0
input_data_mean = 0
anomaly_cut_off = 0
lower_limit = 0
upper_limit = 0
def welford(x, index):
    global input_data_mean
    global input_data_std
    global anomaly_cut_off
    if index == 1:
       input_data_mean=x
       input_data_std=0
       anomaly_cut_off = min(max(input_data_std*3,tolerance), tolerance)
    else:
      M = input_data_mean
      S = input_data_std
      Mnext = M + (x - M) / index
      S = S + (x - M)*(x - Mnext)
      input_data_mean=Mnext
      input_data_std=(S/(index-1))
      anomaly_cut_off = min(max(input_data_std*3,tolerance),tolerance)
    print("mean:" , input_data_mean)
    print("std:" , input_data_std)
    print("cut:" , anomaly_cut_off)

upper = []
def find_anomalies(outlier, index):
    print("candidate outlier:", outlier)
    global lower_limit, upper_limit
    lower_limit  = input_data_mean - anomaly_cut_off 
    upper_limit = input_data_mean + anomaly_cut_off
    print("lower:", lower_limit)
    print("upper:", upper_limit)
    if index!=1 and (outlier > upper_limit or outlier < lower_limit):
        outliers.append(outlier)
        upper.append(upper_limit)
    else:
        clean_data.append(outlier)
        welford(outlier, index)

# build reference statistics for a given window and use it as the initial
# statistics info for valid functioning
train = data[0:1000]
train=train[train>1]

for i in range(0, len(train)):
    welford(train[i], i+1)
print("ANOMALIES Detection...")
for i in range(1000, 2000):
    find_anomalies(data[i],i+1)

#print(outliers)

plt.plot(outliers)
plt.plot(upper)
plt.show()

def adwin_calc(x_array):
   adwin = ADWIN()
   #Adding stream elements to ADWIN and verifying if drift occurred
   for i in range(len(x_array)):
       adwin.add_element(x_array[i])
       if adwin.detected_change():
           print('Change detected in data: ' + str(x_array[i]) + ' - at index: ' + str(i))
print("Cleaned data adwin")
adwin_calc(clean_data)
print("Original data adwin")
adwin_calc(data)

                
    
```
